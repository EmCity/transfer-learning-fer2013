```python
# load data
from preprocessing import *
from constants import *
import matplotlib.pyplot as plt
```

```python
fer_data_train, fer_data_test, fer_label_train, fer_label_test = read_fer()
jaffe_data_train, jaffe_data_test, jaffe_label_train, jaffe_label_test = read_jaffe()
ck_data_train, ck_data_test, ck_label_train, ck_label_test = read_ck()
```

```python
train_data = np.concatenate([fer_data_train, jaffe_data_train, ck_data_train], axis=0)
test_data = np.concatenate([fer_data_test, jaffe_data_test, ck_data_test], axis=0)
train_label = np.concatenate([fer_label_train, jaffe_label_train, ck_label_train], axis=0)
test_label = np.concatenate([fer_label_test, jaffe_label_test, ck_label_test], axis=0)
```
